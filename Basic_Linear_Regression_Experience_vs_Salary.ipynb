Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import RandomOverSampler
import tensorflow as tf
```

Dataset

```python
dataset_cols = ["years_experience","salary"]
df = pd.read_csv("Salary.csv")
df.columns = dataset_cols
df.head()
```

Data Visualization

```python
plt.hist(df["years_experience"])
plt.title("Histogram of Years of Experience")
plt.ylabel("No. of Employees")
plt.xlabel("Years of Experience")
plt.show()
```

```python
plt.hist(df["salary"])
plt.title("Histogram of Salary")
plt.ylabel("No. of Employees")
plt.xlabel("Salary")
plt.show()
```

```python
plt.scatter(df["years_experience"], df["salary"])
plt.title("Years of Experience VS Salary")
plt.ylabel("Salary")
plt.xlabel("Years of Experience")
plt.show()
```

Training the Model

```python
x = df.iloc[:,0].values
y = df.iloc[:,1].values
```

```python
x = x.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
y_pred = regressor.predict(x_test)
y_pred
```

Training data

```python
plt.scatter(x_train, y_train, label="Data", color="blue")
x = tf.linspace(0, 20, 100)
plt.plot(x, regressor.predict(np.array(x).reshape(-1, 1)), label="Fit", color="red", linewidth=3)
plt.legend()
plt.title("Experience vs Salary")
plt.ylabel("Salary")
plt.xlabel("Experience")
plt.show()
```

Testing Data

```python
plt.scatter(x_test, y_test, label="Data", color="blue")
x = tf.linspace(0, 20, 100)
plt.plot(x, regressor.predict(np.array(x).reshape(-1, 1)), label="Fit", color="red", linewidth=3)
plt.legend()
plt.title("Experience vs Salary")
plt.ylabel("Salary")
plt.xlabel("Experience")
plt.show()
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
user_input = float(input("Enter years of experience: "))
years_of_experience = float(user_input)
predicted_salary = regressor.predict([[years_of_experience]])[0]
print("Predicted salary:", predicted_salary)
```
